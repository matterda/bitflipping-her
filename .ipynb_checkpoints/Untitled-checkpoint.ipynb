# Bit-flipping problem
As described in https://arxiv.org/abs/1707.01495

```python
import numpy as np
import matplotlib.pyplot as plt

from BitFlipping_single_episode import BitFlipEnv, DQN
```

```python
def train(N, epochs, gamma, buffer_size, use_her):
    env = BitFlipEnv(N)
    model = DQN(env, gamma, buffer_size)
    model.use_her = use_her
    start_obs, done = env.reset()
    
    for e in range(epochs):
        model.episode()
    dist = np.asarray(model.logger.get_log('final_dist'))
    done = (dist == 0)
    wind_size = 50
    succ_rate = done.reshape(-1, wind_size).sum(1)/wind_size
    return model, dist, {'x': np.arange(epochs/wind_size)*wind_size, 'y': succ_rate}
```

```python
N = 20
epochs = 5000
gamma = 0.98
buffer_size = 100000

_, _, succ_rate = train(N, epochs, gamma, buffer_size, True)
plt.plot(succ_rate['x'], succ_rate['y'],label='DQN+HER')

_, _, succ_rate = train(N, epochs, gamma, buffer_size, False)
plt.plot(succ_rate['x'], succ_rate['y'],label='DQN')

plt.title(f'N = {N}')
plt.xlabel('Epoch')
plt.ylabel('Success rate')
plt.legend()
plt.show()
```
